```python
# first corrections --> looking for missing values, then deciding how to handle them, replace at random, through a certain method, drop, etc.
# alligning gender names
d['Gender'] = d['Gender'].str.replace('Male', 'male')
d['Gender'] = d['Gender'].str.replace('Female', 'female')
d['Gender'] = d['Gender'].str.replace('FeMale', 'female')
```

```python
# drop categories --> unusuable column
del d['Var_1']
```

```python
# updating graduates
# unique Professions
lst1 = set(d['Profession'].tolist())        
lst1
prof_list = ['Doctor', 'Lawyer']
for x in prof_list:
    d.loc[d['Profession'] == x, 'Graduated'] = 'Yes'
    
```

```python
# updating work experience

mean = d['Work_Experience'].mean()
```

```python
std = d['Work_Experience'].std()
```

```python
sizenan = d['Work_Experience'].isnull().sum()
```

```python

rand = np.random.randint(mean - 2*std, mean + 2*std, size = sizenan)
d["Work_Experience"][np.isnan(d["Work_Experience"])] = rand
```

```python
# filling up the rest of graduated missing values, used the ffill method because percentage is so low
d['Graduated'] = d['Graduated'].ffill()
```

```python
# updating Profession based on the input spending score

d['Spending_Score'].unique()
```

```python
def sc_encode(uniquesc, value):   
    d.loc[d['Spending_Score'] == f'{uniquesc}', 'Spending_Score_encoded'] = value

sc_encode('Low', 0)
sc_encode('Average', 1)
sc_encode('High', 2)
```

```python
profsc = d.groupby('Profession', as_index=False).agg({'Spending_Score_encoded':'mean'})
profsc.round()
```

```python
d['Profession'].replace(np.nan, "no", inplace=True)
```

```python
import random
list_1 = ['Artist','Executive','Lawyer']        
list_0 = ['Doctor','Engineer','Entertainment','Healthcare','Homemaker','Marketing']
list_2 = ['Executive', 'Lawyer']

x = random.choice(list_1)
y = random.choice(list_0)
c = random.choice(list_2)

d.loc[(d['Spending_Score_encoded'] ==1)&(d["Profession"] =="no"),'Profession'] = x
d.loc[(d['Spending_Score_encoded'] ==0)&(d["Profession"] =="no"),'Profession'] = y
d.loc[(d['Spending_Score_encoded'] ==2)&(d["Profession"] =="no"),'Profession'] = c

```

```python
d.head(60)
```

```python
# ever married
d['Ever_Married'] = d['Ever_Married'].ffill()
```

```python
# updating Falimiy_size
mean = d['Family_Size'].mean()
std = d['Family_Size'].std()
sizenan = d['Family_Size'].isnull().sum()
```

```python
rand = np.random.randint(mean - 2*std, mean + 2*std, size = sizenan)
d["Family_Size"][np.isnan(d["Family_Size"])] = rand
```
